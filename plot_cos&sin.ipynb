```python
# coding: utf-8
import numpy as np
import matplotlib.pyplot as plt

class plot10530041:
    """functions of class """

    def __init__(self,x):
        self.x = np.arange(0, 6, 0.1)


    def sin(self):
      
        y1 = np.sin(self.x)
        
        
        plt.plot(self.x, y1, label="sin")
        
        plt.xlabel("x") # x軸
        plt.ylabel("y") # y軸
        plt.title('sin')
        plt.legend()
        plt.show()




    def cos(self):
        
        y2 = np.cos(self.x)
        plt.plot(self.x, y2, linestyle = "--", label="cos")
        plt.xlabel("x") # x軸
        plt.ylabel("y") # y軸
        plt.title('cos')
        plt.legend()
        plt.show()
        
        
   




p=plot10530041("")
p.sin()
p.cos()
```
